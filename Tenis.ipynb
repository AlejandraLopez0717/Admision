## Tenis
El tenis tiene un sistema de puntuación bastante peculiar, y para los recién llegados puede ser un poco difícil seguirle la pista. La sociedad de tenis lo ha contratado para que construya un marcador para mostrar el puntaje actual durante los juegos de tenis.

Puede leer más sobre las puntuaciones de tenis en [wikipedia](https://en.wikipedia.org/wiki/Tennis#Scoring), que se resumen a continuación:

1.   Un juego lo gana el primer jugador que ha ganado al menos cuatro puntos en total y al menos dos puntos más que el oponente.
2.   La puntuación de ejecución de cada juego se describe de una manera peculiar del tenis: las puntuaciones de cero a tres puntos se describen como "Cero", "Quince", "Treinta" y "Cuarenta" respectivamente.
3. Si cada lado ha anotado al menos tres puntos y un jugador tiene un punto más que su oponente, el puntaje del juego es "Ventaja" para el jugador que va en cabeza.


Solo necesita informar la puntuación del juego actual. Los conjuntos y los partidos están fuera del alcance.


```python
class Tennis:
    SCORE = {0: "Cero", 1: "Quince", 2: "Treinta", 3: "Cuarenta"}

    def __init__(self):
        self.score = {"Jugador 1": 0, "Jugador 2": 0}

    def add_score(self, jugador):
        if jugador in self.score:
            self.score[jugador] += 1
        else:
            print(f"{jugador} no es un jugador válido.")

    def show_score(self):
        if "Jugador 1" in self.score and "Jugador 2" in self.score:
            score1 = self.score["Jugador 1"]
            score2 = self.score["Jugador 2"]

            if score1 == score2:
                if score1 < 3:
                    print(f"{self.SCORE[score1]} iguales")
                else:
                    print("Empate")
            elif score1 >= 4 or score2 >= 4:
                if abs(score1 - score2) >= 2:
                    if score1 > score2:
                        print("Jugador 1 gana el juego")
                    else:
                        print("Jugador 2 gana el juego")
                else:
                    if score1 > score2:
                        print("Ventaja Jugador 1")
                    else:
                        print("Ventaja Jugador 2")
            else:
                print(f"{self.SCORE[score1]} - {self.SCORE[score2]}")
        else:
            print("No se pueden mostrar puntajes si los jugadores no están definidos.")

    def finish_score(self):
        while True:
            jugador = input("Ingrese el número del jugador que anotó (1 o 2) o 'fin' para terminar: ")
            if jugador == "fin":
                break
            elif jugador in ["1", "2"]:
                jugador_nombre = f"Jugador {jugador}"
                self.add_score(jugador_nombre)
                self.show_score()

scoreboard = Tennis()
scoreboard.finish_score()
```
